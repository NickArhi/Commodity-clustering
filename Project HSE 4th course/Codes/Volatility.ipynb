```python
from arch import arch_model

def apply_garch_model(df):
    cond_vol_dict = {}

    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            model = arch_model(df[column], vol='Garch', p=1, o=1, q=1, dist='StudentsT')
            res = model.fit(update_freq=5, disp='off')
            cond_vol_dict[column] = res.conditional_volatility


    cond_vol_df = pd.DataFrame(cond_vol_dict)
    return cond_vol_df

conditional_vol_df = apply_garch_model(return_log_scaled)
conditional_vol_df.head()
```

```python
from arch import arch_model


columns_to_model = return_log_scaled.columns.drop('Date')

best_models = {}

for column in columns_to_model:
    print(f"Fitting GJR-GARCH model for {column}")
    series = return_log_scaled[column].dropna()  

    best_aic = float('inf')
    best_params = None

    for p in range(1, 6):
        for o in range(0, 6):
            for q in range(1, 6):
                try:
                    model = arch_model(series, p=p, o=o, q=q, vol='Garch', dist='StudentsT')
                    res = model.fit(disp='off')
                    if res.aic < best_aic:
                        best_aic = res.aic
                        best_params = (p, o, q)
                except Exception as e:
                    print(f"An error occurred while fitting model for {column} with p={p}, o={o}, q={q}: {str(e)}")
                    continue

    print(f"Best parameters for {column}: p={best_params[0]}, o={best_params[1]}, q={best_params[2]}")

    best_model = arch_model(series, p=best_params[0], o=best_params[1], q=best_params[2], vol='Garch', dist='StudentsT')
    fitted_model = best_model.fit(disp='off')
    best_models[column] = fitted_model

    print(fitted_model.summary())
```
